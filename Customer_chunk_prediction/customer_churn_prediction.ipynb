```python
df1= df[df.TotalCharges!=' ']
df1.shape
```

```python
df1.dtypes
```

```python
#now we can convert into numeric values still it was in string na 
df1.TotalCharges=pd.to_numeric(df1.TotalCharges)
```

```python
#how many customer are leaving x axis can be tenure
tenure_churn_no=df1[df1.Churn=='No'].tenure
tenure_churn_yes=df1[df1.Churn=='Yes'].tenure

plt.xlabel("Tenure")
plt.ylabel("NO. of Customers")
plt.title('Cutomer Churn Prediction Visuallisation')

plt.hist([tenure_churn_yes, tenure_churn_no], color=['green','red'], label=['churnyes','churnNo'])
plt.legend()
```

```python
#same thing for monthly charges at place of tenure
#how many customer are leaving x axis can be monthly charges
MonthlyCharges_churn_no=df1[df1.Churn=='No'].MonthlyCharges
MonthlyCharges_churn_yes=df1[df1.Churn=='Yes'].MonthlyCharges

plt.xlabel("MonthlyCharges")
plt.ylabel("NO. of Customers")
plt.title('Cutomer Churn Prediction Visuallisation')

plt.hist([MonthlyCharges_churn_yes, MonthlyCharges_churn_no], color=['green','red'], label=['MonthlyChargesYes','MonthlyChargesNo'])
plt.legend()
```

```python

def print_unique_col_values(df):
    for column in df:
        if df[column].dtypes=='object':
            print(f'{column} : {df[column].unique()}')
```

```python
print_unique_col_values(df1)
```
